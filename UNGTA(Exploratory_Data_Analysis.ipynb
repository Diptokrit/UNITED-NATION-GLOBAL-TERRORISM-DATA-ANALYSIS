### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

#just to be safe
df_original = df.copy()

#the dataset is old and it has been treating -99 as a placeholder for the misssing vals
df_eda = df.replace(-99,np.nan, inplace = True)


df

````

````python
#rechecking for any missing vals

first_10 = df.isnull().sum().sort_values(ascending=False)
first_10.head(50)
````

After replacing placeholder vals (-99) with nan , the missing value distribution was re-evaluated . It was observed that several columns contain a very high number of missing values , in some cases nearly entire dataset .

these columns are not usefull for the analysis and can be dropped for EDA

At the same time some some columns contain no missing values and are reliable for the further analysis

````python
df_eda = df[['iyear', 'imonth','country_txt','region_txt','city', 'attacktype1_txt','targtype1_txt','weaptype1_txt',
             'gname','nkill','nwound']].copy()
````

**Temporal variables** : iyear , imonth

**Location based Insights** : country_txt, region_txt, city


**Nature of the attack** : attacktype1_txt, targtype1_txt, weaptype1_txt

**Groups involved** : gname

**Severe effects** : nkill, nwound


Now we will perform data cleaningbased on just these columns specifically

````python
df_eda.isnull().sum()
````

````python
df_eda['nkill'] = df_eda['nkill'].fillna(0)
df_eda['nwound'] = df_eda['nwound'].fillna(0)
````

````python
df_eda['city'] = df_eda['city'].fillna('Unknown')
````

Correlation heatmaps are useful for identifying patterns and dependencies between variables in a compact visual format.

##### 2. What is/are the insight(s) found from the chart?

This indicates that incidents with higher fatalities tend to also result in a higher number of injuries. However, the correlation is not very strong, suggesting that the number of deaths and injuries can vary depending on the nature and intensity of the attack.
